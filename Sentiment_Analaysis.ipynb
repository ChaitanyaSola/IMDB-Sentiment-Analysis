```python
#Import Libraries
import numpy as np
import joblib
from tensorflow.keras.datasets import imdb
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
#Load IMDB Dataset
#loadad top 10,000 most common words.

num_words = 10000

(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=num_words)

print("Train samples:", len(X_train))
print("Test samples:", len(X_test))
print("Unique classes:", np.unique(y_train))
```

```python
#Convert Integer Sequences Back to Text
#IMDB dataset gives numbers, not text.
#We must decode them.

word_index = imdb.get_word_index()
reverse_word_index = {value: key for key, value in word_index.items()}

def decode_review(encoded_review):
    return " ".join(
        [reverse_word_index.get(i - 3, "?") for i in encoded_review]
    )

X_train_text = [decode_review(review) for review in X_train]
X_test_text = [decode_review(review) for review in X_test]
```

```python
#TF-IDF Vectorization
vectorizer = TfidfVectorizer(stop_words="english", max_features=5000)

X_train_vec = vectorizer.fit_transform(X_train_text)
X_test_vec = vectorizer.transform(X_test_text)
```

```python
#Train Logistic Regression
model = LogisticRegression(max_iter=1000)
model.fit(X_train_vec, y_train)
```

```python
#Evaluate Model
y_pred = model.predict(X_test_vec)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
joblib.dump(model, "model.pkl")
joblib.dump(vectorizer, "vectorizer.pkl")
```
